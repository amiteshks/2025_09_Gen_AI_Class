```python
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Load the document
file_path = "data/Motor_Vehicle_Claim_Complex.pdf"
loader = PyPDFLoader(file_path)

# This creates a list of Document objects, one per page
docs = loader.load()
print(f"Loaded {len(docs)} pages.")

# Create a text splitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=400,      # number of characters per chunk
    chunk_overlap=100,   # overlap to preserve context
    separators=["\n\n", "\n", ".", " ", ""],  # split by structure first
)

# Split the documents
chunks = text_splitter.split_documents(docs)
print(f"Created {len(chunks)} chunks.")

# Inspect the first few chunks
for i, chunk in enumerate(chunks):
    print(f"\n--- Chunk {i+1} ---")
    print(chunk.page_content[:400])  # preview first 400 characters
    print("\nMetadata:", chunk.metadata)
```
